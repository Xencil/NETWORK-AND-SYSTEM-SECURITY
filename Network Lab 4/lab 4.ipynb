```python
import csv
import os
import hashlib
from datetime import datetime

path = r"C:\Users\jaris\Downloads\NETWORK AND SYSTEM SECURITY\Network Lab 4"
externalFile = os.path.join(path, "fileHashes.csv")
CheckerFile = os.path.join(path, "newFileHashes.csv")


def hash(file_path):
    sha256 = hashlib.sha256()
    with open(file_path, "rb") as f:
        for byte in iter(lambda: f.read(4096), b""):
            sha256.update(byte)
    return sha256.hexdigest()

with open(CheckerFile, mode="w", newline="", encoding="utf-8") as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(["file", "sha-256 hash", "time"])
    filesDone = 0
    for root, dirs, files in os.walk(path):
        for file in files:
            if file in [os.path.basename(externalFile), os.path.basename(CheckerFile)]:
                continue
            file_path = os.path.join(root, file)
            file_hash = hash(file_path)
            time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            writer.writerow([file_path, file_hash, time])
            filesDone += 1

def load_hashes(csv_path):
    hashes = {}
    with open(csv_path, newline="", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        for row in reader:
            hashes[row["file"]] = row["sha-256 hash"]
    return hashes
baseline_hashes = load_hashes(externalFile)
new_hashes = load_hashes(CheckerFile)

Changed = []
deleted =[]
added = []

for file,old_hash in baseline_hashes.items():
    if file not in new_hashes:
        deleted.append(file)
    elif new_hashes[file] != old_hash:
        Changed.append(file)

for file in new_hashes:
    if file not in baseline_hashes:
        added.append(file)

if Changed:
    print("changed files:")
    for x in Changed:
        print(x)
else:
    print("No chnages")

if deleted:
    print("Deleted files:")
    for x in deleted:
        print(x)
else:
    print("No deleted files")

if added:
    print("New added files")
    for x in added:
        print(x)
else:
    print("No new files")
```

```python

import re
from pathlib import Path

SIGNATURES = [r"eval\(", r"base64\.b64decode", r"socket\.connect", r"exec\(", r"import os"]
PATTERNS = [(s, re.compile(s, re.IGNORECASE)) for s in SIGNATURES]

def scan_file(p):
    try:
        with open(p, "r", encoding="utf-8", errors="replace") as f:
            for n, line in enumerate(f, 1):
                for s, rx in PATTERNS:
                    m = rx.search(line)
                    if m:
                        print(f"{p}:{n} [{s}] -> {m.group(0).strip()}")
    except Exception:
        pass 

def scan_folder(folder):
    for p in Path(folder).rglob("*"):
        if p.is_file():
            scan_file(p)

if __name__ == "__main__":
    here = Path(".").resolve()
    print(f"Scanning folder: {here}")
    scan_folder(here)
```

```python
import random
InfectCount =set([0]) 
Devices = 30
Step =20
Rate =3
Chance = 0.4

for step in range(Step):
    New =set()
    for i in InfectCount:
        for x in range(Rate):
            Target =random.randint(0,Devices- 1)
            if Target not in InfectCount:
                if random.random()<Chance:
                    New.add(Target)
    InfectCount |= New
    print(f"{step +1}: infected {len(InfectCount)} device")
    if len(InfectCount)==Devices:
        break

```

```python
Threshold = 5
Log = []
InfectCount = set([0])
Connections = {host: 0 for host in range(Devices)}


def AlertLog(step, host, connections):
    time =datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    msg =f"[ALERT] Step {step}: {time} host {host} has {connections} connections"
    Log.append(msg)
    print(msg)
for i in range(1, Step + 1):
    print(f"\nStage{i}")
    new =set()
    for x in InfectCount:
        for x in range(Rate):
            target = random.randint(0, Devices - 1)
            Connections[target] += 1
            if random.random() < Chance:
                new.add(target)
    InfectCount |= new
    for y in range(Devices):
        connections =Connections[y]
        bar ='#'*connections
        print(f"Host {y:02}: {bar} ({connections})", end=' ')
        if connections>Threshold:
            print("-", end=' ')
            AlertLog(i, y, connections)
        print()

if Log:
    for msg in Log:
        print(msg)
else:
    print("No anomalies detected.")
```
