```python
import pathlib
import os
from IPython.display import display, HTML
Link ="https://google-gruyere.appspot.com/367371353373946205459788452183384869795/"
Folder = pathlib.Path(os.getcwd()).resolve()
repFolder =Folder / "reports"
repFolder.mkdir(exist_ok=True)
report ="juice_wapiti_report.html"
reportPath = repFolder/report

if reportPath.is_file():
    found =reportPath
elif reportPath.is_dir():
    fileH = list(reportPath.glob("*.html"))
    if fileH:
        fileH.sort(key=lambda p:p.stat().st_mtime,reverse= True)
        found =fileH[0]
    else:
        found =None
else:
    found=None

if found and found.exists():
    Htext =found.read_text(encoding="utf-8",errors="replace")
    display(HTML(Htext))

```
